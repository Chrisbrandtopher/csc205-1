```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../images/pair-color.png")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
def imshow(img, title=""):
    plt.figure(figsize=(8, 8))
    plt.imshow(img, cmap="gray")
    _ = plt.yticks([])
    _ = plt.xticks([])
    plt.title(title)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray, cmap="gray")
plt.yticks([])
plt.xticks([])
```

```python
img_disp = np.maximum(0, np.minimum(255, img_gray.astype(int)))
plt.figure(figsize=(8, 8))
plt.imshow(img_disp, cmap="gray")
_ = plt.yticks([])
_ = plt.xticks([])
```

```python
# Filtering example 1 (mean filter)
filter = 1/9*np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1],
])
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
plt.figure(figsize=(8, 8))
plt.imshow(img_disp, cmap="gray")
_ = plt.yticks([])
_ = plt.xticks([])
```

```python
# Filtering example 1 (mean filter)
filter = 1/(16**2)*np.ones((16,16))
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
plt.figure(figsize=(8, 8))
plt.imshow(img_disp, cmap="gray")
_ = plt.yticks([])
_ = plt.xticks([])
```

```python
# Boundary example with mean filter, border constant
N = 64
filter = 1/(N**2)*np.ones((N,N))
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter, borderType=cv2.BORDER_CONSTANT)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res)
```

```python
# Boundary example with mean filter, border repeat
N = 64
filter = 1/(N**2)*np.ones((N,N))
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter, borderType=cv2.BORDER_REPLICATE)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res)
```

```python
# Boundary example with mean filter, border reflect
N = 64
filter = 1/(N**2)*np.ones((N,N))
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter, borderType=cv2.BORDER_REFLECT)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res)
```

```python
# Boundary example with mean filter, border cycle
N = 64
filter = 1/(N**2)*np.ones((N,N))
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter, borderType=cv2.BORDER_WRAP)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res)
```

```python
# Boundary example with mean filter, border reflect 101
N = 64
filter = 1/(N**2)*np.ones((N,N))
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter, borderType=cv2.BORDER_REFLECT_101)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res)
```

```python
# Example image for box
img_box = np.ones((520, 780), dtype=np.uint8) * 255
img_box[128:(520-128), 256:(780-256)] = 0
imshow(img_box)
```

```python
#  Mean filter blur example for pear image
N = 65
filter = 1/(N**2)*np.ones((N,N))
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter, borderType=cv2.BORDER_REFLECT_101)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res)
```

```python
# Mean filter blur example for box image
N = 65
filter = 1/(N**2)*np.ones((N,N))
img_res = cv2.filter2D(img_box, cv2.CV_32F, filter, borderType=cv2.BORDER_REFLECT_101)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res)
```

```python
# Gaussian filter example
filter1D = cv2.getGaussianKernel(65, 11)
filterG = filter1D * filter1D.T
filterG /= filterG.sum()
imshow(filterG, "65 x 65 Gaussian Kernel with Std = 11")
```

```python
# Gaussian blur example
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filterG, borderType=cv2.BORDER_REFLECT_101)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res, "Blurred with Gaussian Kernel")
```

```python
N = 31
filterM = 1/(N**2)*np.ones((N,N))
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filterM, borderType=cv2.BORDER_REFLECT_101)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res, "Blur with Mean Filter")
```

```python
img_res = cv2.filter2D(img_box, cv2.CV_32F, filterG, borderType=cv2.BORDER_REFLECT_101)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res, "Blur with Gaussian Kernenl")
```

```python
img_res = cv2.filter2D(img_box, cv2.CV_32F, filterM, borderType=cv2.BORDER_REFLECT_101)
img_disp = np.maximum(0, np.minimum(255, img_res.astype(int)))
imshow(img_res, "Blur with Mean FIlter")
```

```python
# Gaussian filter example
std = 5
filter1D = cv2.getGaussianKernel(65, std)
filterG = filter1D * filter1D.T
filterG /= filterG.sum()
imshow(filterG, "65 x 65 Gaussian Kernel with Std = {}".format(std))
```

```python
# Gaussian filter example
std = 31
filter1D = cv2.getGaussianKernel(65, std)
filterG = filter1D * filter1D.T
filterG /= filterG.sum()
imshow(filterG, "65 x 65 Gaussian Kernel with Std = {}".format(std))
```

```python
# Sharpening
imshow(img_gray, "Original")
smoothed = cv2.filter2D(img_gray, cv2.CV_32F, filterG, borderType=cv2.BORDER_REFLECT_101)
detail = img_gray - smoothed
imshow(detail, "Detail")
sharpened = img_gray + 1.0 * detail
img_disp = np.maximum(0, np.minimum(255, sharpened.astype(int)))
imshow(img_disp, "Sharpened Image")
```

```python
# Laplacian of Gaussian Filter
from scipy.ndimage import gaussian_laplace
filter_log = np.zeros((65, 65), dtype=np.float)
filter_log[32,32] = 1.0
filter_log = gaussian_laplace(filter_log, 3)
filter_log = filter_log
imshow(filter_log * 255, "LoG Filter")

# Filtering original image with this filter
imshow(img_gray, "Original")
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter_log, borderType=cv2.BORDER_REFLECT_101)
imshow(img_res, "LoG Filtered")

```

```python
# Taking the absolute and thresholding
filter_log = np.zeros((65, 65), dtype=np.float)
filter_log[32,32] = 1.0
filter_log = gaussian_laplace(filter_log, 5)
filter_log = filter_log
img_res = cv2.filter2D(img_gray, cv2.CV_32F, filter_log, borderType=cv2.BORDER_REFLECT_101)
imshow(abs(img_res), "LoG Filtered (abs)")

imshow(abs(img_res) > 0.4, "LoG Filtered (abs) thresholded")
```
