```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../images/pair-color.png")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
# Histogram
plt.figure(figsize=(5, 2))
_, _, _ = plt.hist(img_gray.flatten(), bins=np.linspace(0, 255, 255), color="black")
plt.yticks([])
plt.xticks([0, 255])
plt.xlim(0, 255)
```

```python
# Cumulative Histogram
plt.figure(figsize=(5, 2))
_, _, _ = plt.hist(img_gray.flatten(), bins=np.linspace(0, 255, 255), color="black", cumulative=True)
plt.yticks([])
plt.xticks([0, 255])
plt.xlim(0, 255)
```
