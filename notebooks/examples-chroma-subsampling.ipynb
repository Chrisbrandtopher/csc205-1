```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(img, title=""):
    plt.figure(figsize=(8, 8))
    plt.imshow(img, cmap="gray", interpolation="lanczos")
    _ = plt.yticks([])
    _ = plt.xticks([])
    plt.title(title, fontsize=20)
```

```python
img = cv2.imread("../images/fruit.jpg")[...,::-1]
```

```python
imshow(img, "Fruit")
```

```python
# Down sample and upsample image to emulate storing low-resolution image per channel
ds_rate = 5
img_ds = cv2.resize(img, (img.shape[1]//ds_rate, img.shape[0]//ds_rate), interpolation=cv2.INTER_NEAREST)
img_us = cv2.resize(img_ds, img.shape[:2][::-1], interpolation=cv2.INTER_NEAREST)
# Create a new image using only the green channel for the full resolution one
img_ss = img_us.copy()
img_ss[:,:,1] = img[:,:,1]
imshow(img_ss, "Fruit, {} times downsampled (NN)".format(ds_rate))
```

```python
ds_rate = 10
img_ds = cv2.resize(img, (img.shape[1]//ds_rate, img.shape[0]//ds_rate), interpolation=cv2.INTER_NEAREST)
img_us = cv2.resize(img_ds, img.shape[:2][::-1], interpolation=cv2.INTER_NEAREST)
# Create a new image using only the green channel for the full resolution one
img_ss = img_us.copy()
img_ss[:,:,1] = img[:,:,1]
imshow(img_ss, "Fruit, {} times downsampled (NN)".format(ds_rate))
```

```python
ds_rate = 10
img_ds = cv2.resize(img, (img.shape[1]//ds_rate, img.shape[0]//ds_rate))
img_us = cv2.resize(img_ds, img.shape[:2][::-1])
# Create a new image using only the green channel for the full resolution one
img_ss = img_us.copy()
img_ss[:,:,1] = img[:,:,1]
imshow(img_ss, "Fruit, {} times downsampled (Linear)".format(ds_rate))
```

```python
ds_rate = 10
img_ds = cv2.resize(img, (img.shape[1]//ds_rate, img.shape[0]//ds_rate))
img_us = cv2.resize(img_ds, img.shape[:2][::-1])
# Create a new image using only the green channel for the full resolution one
img_ss = img_us.copy()
img_ss[:,:,0] = img[:,:,0]
imshow(img_ss, "Subsampling by 10, R channel preserved".format(ds_rate))
img_ss = img_us.copy()
img_ss[:,:,1] = img[:,:,1]
imshow(img_ss, "Subsampling by 10, G channel preserved".format(ds_rate))
img_ss = img_us.copy()
img_ss[:,:,2] = img[:,:,2]
imshow(img_ss, "Subsampling by 10, B channel preserved".format(ds_rate))
```
