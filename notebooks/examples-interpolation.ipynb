# Examples for interpolation

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(img, title=""):
    plt.figure(figsize=(4, 4), dpi=192)
    # This one is just for display!
    plt.imshow(img, cmap="gray", vmin = 0, vmax = 255, interpolation="nearest") 
    _ = plt.yticks([])
    _ = plt.xticks([])
    plt.title(title, fontsize=10)
```

```python
# Load image
img_gray = cv2.imread("../images/sign.png", 0)
img_gray = cv2.resize(img_gray, (200, 200))

# Show image
imshow(img_gray, "Original image (200 x 200)")
```

```python
# Resize image to be smaller
img_resize = cv2.resize(img_gray, (120, 120), interpolation=cv2.INTER_NEAREST)
imshow(img_resize, "Resized Image (120 x 120) -- NN")

img_resize = cv2.resize(img_gray, (120, 120), interpolation=cv2.INTER_LINEAR)
imshow(img_resize, "Resized Image (120 x 120) -- Bilinear")

img_resize = cv2.resize(img_gray, (120, 120), interpolation=cv2.INTER_CUBIC)
imshow(img_resize, "Resized Image (120 x 120) -- Bicubic")

img_resize = cv2.resize(img_gray, (120, 120), interpolation=cv2.INTER_LANCZOS4)
imshow(img_resize, "Resized Image (120 x 120) -- Lanczos")
```
