```python
# Numpy library to use matrices
import numpy as np
import cv2

# Load display library for jupyter notebook
import matplotlib
from matplotlib import pyplot as plt
```

```python
# We'll continue onwards with our canvas function
N = 100

def show_canvas(canvas, title=""):
    # Similar to doing cv2.imshow, but in this environment
    plt.figure(figsize = (8,8))
#     plt.axis('off')
    plt.imshow(canvas, cmap="gray", norm=matplotlib.colors.NoNorm()) 
    plt.title(title)
    
def reset_canvas(size):
    return np.ones((size,size), dtype=np.uint8) * 255
    
```

```python
canvas = reset_canvas(N)
show_canvas(canvas)
```

```python
# Draw a line
canvas = reset_canvas(N)
canvas = cv2.line(canvas, (20,20), (80, 43), 0, 1, cv2.LINE_AA)
show_canvas(canvas)
```

```python
# Draw a circle

canvas = reset_canvas(N)
canvas = cv2.circle(canvas, (40,40), 3, 0, 1, cv2.LINE_AA)
canvas = cv2.circle(canvas, (40,40), 10, 0, 1, cv2.LINE_AA)
show_canvas(canvas)
```

```python
# Draw triangle
x0 = (10,10)
x1 = (50,50)
x2 = (90,10)

X = np.array([[x0, x1, x2, x0]], np.int32)

canvas = reset_canvas(N)
canvas = cv2.fillPoly(canvas, X, 0, cv2.LINE_AA)
show_canvas(canvas)
```

```python
# Draw polygons
X = np.array([[
    [10, 10], 
    [50, 50], 
    [90, 10], 
    [10, 30], 
    [10, 90], 
    [90, 90], 
    [90, 50], 
    [10, 10], 
    ]], np.int32)

canvas = reset_canvas(N)
canvas = cv2.fillPoly(canvas, X, 0, cv2.LINE_AA)
show_canvas(canvas)
```

```python
# Draw polygons
X = np.array([[
    [10, 10], 
    [50, 50], 
    [90, 10], 
    [10, 30], 
    [10, 90], 
    [90, 90], 
    [90, 50], 
    [10, 10], 
    ]], np.int32)

canvas = reset_canvas(N)
for i in range(len(X[0])):
    canvas = cv2.line(canvas, tuple(X[0][i-1]), tuple(X[0][i]), 0, 1, cv2.LINE_AA)
show_canvas(canvas)
```
