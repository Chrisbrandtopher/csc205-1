```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../images/pair-color.png")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray, cmap="gray")
plt.yticks([])
plt.xticks([])
```

```python
img_disp = np.maximum(0, np.minimum(255, img_gray.astype(int)))
plt.figure(figsize=(8, 8))
plt.imshow(img_disp, cmap="gray")
plt.yticks([])
plt.xticks([])
plt.figure(figsize=(8, 3))
_, _, _ = plt.hist(img_disp.flatten(), bins=np.linspace(0, 255, 255), color="black")
plt.yticks([])
plt.xticks([0, 255])
plt.xlim(0, 255)
```

```python
img_disp = np.maximum(0, np.minimum(255, img_gray.astype(int) * 0.7))
plt.figure(figsize=(8, 8))
plt.imshow(img_disp, cmap="gray")
plt.yticks([])
plt.xticks([])
plt.figure(figsize=(8, 3))
_, _, _ = plt.hist(img_disp.flatten(), bins=np.linspace(0, 255, 255), color="black")
plt.yticks([])
plt.xticks([0, 255])
plt.xlim(0, 255)
```

```python
img_disp = np.maximum(0, np.minimum(255, img_gray.astype(int) - 100))
plt.figure(figsize=(8, 8))
plt.imshow(img_disp, cmap="gray")
plt.yticks([])
plt.xticks([])
plt.figure(figsize=(8, 3))
_, _, _ = plt.hist(img_disp.flatten(), bins=np.linspace(0, 255, 255), color="black")
plt.yticks([])
plt.xticks([0, 255])
plt.xlim(0, 255)
```

```python
img_disp = np.maximum(0, np.minimum(255, img_gray.max() - img_gray))
plt.figure(figsize=(8, 8))
plt.imshow(img_disp, cmap="gray")
plt.yticks([])
plt.xticks([])
plt.figure(figsize=(8, 3))
_, _, _ = plt.hist(img_disp.flatten(), bins=np.linspace(0, 255, 255), color="black")
plt.yticks([])
plt.xticks([0, 255])
plt.xlim(0, 255)
```

```python
img_disp = np.maximum(0, np.minimum(255, 255*(img_gray > 90)))
plt.figure(figsize=(8, 8))
plt.imshow(img_disp, cmap="gray")
plt.yticks([])
plt.xticks([])
plt.figure(figsize=(8, 3))
_, _, _ = plt.hist(img_disp.flatten(), bins=np.linspace(0, 255, 255), color="black")
plt.yticks([])
plt.xticks([0, 255])
plt.xlim(0, 255)
```
